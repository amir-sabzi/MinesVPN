```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
datasets_hardstamp = {}
datasets_softstamp = {}
keys = []
intervals = [10,20,50,100,200,500,1000,2000,5000,10000,20000,50000,100000]
for interval in intervals:
    interval_str = str(interval)
    key = interval_str + 'us'
    keys.append(key)
    file_path_hard = glob.glob('/home/asabzi/data/4-exp/40g-NIC/hardstamp/interval_' + interval_str + 'us*.csv')[0]
    file_path_soft = glob.glob('/home/asabzi/data/4-exp/40g-NIC/softstamp/interval_' + interval_str + 'us*.csv')[0]

    value_hard = pd.read_csv(file_path_hard,names=['softstamp', 'dummy', 'hardstamp'], index_col=None)
    value_soft = pd.read_csv(file_path_soft,names=['softstamp', 'dummy', 'hardstamp'], index_col=None) 
    value_hard['hardstamp']=value_hard['hardstamp'].str[:-1].str.replace('.','')
    value_soft['softstamp']=value_soft['softstamp'].str[:-1].str.replace('.','')
    datasets_hardstamp[key] = value_hard
    datasets_softstamp[key] = value_soft

'''df = datasets['10000us']
df_rtx_1 = df.iloc[:-1,[2]].reset_index()
df_rtx_2 = df.iloc[1:,[2]].reset_index()
df_diffs = df_rtx_2['hardstamp'].astype('float64')- df_rtx_1['hardstamp'].astype('float64')
fig = plt.figure()'''

for key in keys:
    df_soft = datasets_softstamp[key]
    df_hard = datasets_hardstamp[key]
    df_rtx_1_hard = df_hard.iloc[:-1,[2]].reset_index()
    df_rtx_2_hard = df_hard.iloc[1:,[2]].reset_index()

    df_rtx_1_soft = df_soft.iloc[:-1,[0]].reset_index()
    df_rtx_2_soft = df_soft.iloc[1:,[0]].reset_index()

    df_diffs_hard = df_rtx_2_hard['hardstamp'].astype('float64')- df_rtx_1_hard['hardstamp'].astype('float64')
    df_diffs_soft = df_rtx_2_soft['softstamp'].astype('float64')- df_rtx_1_soft['softstamp'].astype('float64')

    smean_hard = df_diffs_hard.mean()/1000.0
    sstd_hard = df_diffs_hard.std()/1000.0 
    smean_soft = df_diffs_soft.mean()/1000.0
    sstd_soft = df_diffs_soft.std()/1000.0 

    nominal_interval = float(key[:-2])
    print("----------------------------------")
    print("The average of intervals for %s is: [hard: %.3f, soft: %.3f] us" % (key, smean_hard, smean_soft))
    print("The std of intervals for %s is: [hard: %.3f, soft: %.3f] us" % (key, sstd_hard, sstd_soft))
    print("Time drift percentage is: [hard: %.3f, soft: %.3f]" % (abs(smean_hard - nominal_interval)/nominal_interval * 100,abs(smean_soft - nominal_interval)/nominal_interval * 100) + "%")



```
